```python
import numpy as np
import pandas as pd

from lc_classifier.classifier.models import HierarchicalRandomForest
```

# Loading data

```python
features = pd.read_parquet("data/features_sample.parquet")
```

```python
features.head()
```

# Creating model 

```python
model = HierarchicalRandomForest({})
```

# Loading pre-trained model

```python
# The container comes with the pre-trained model downloaded in a specific path
print(model.MODEL_PICKLE_PATH)
# To load the models we call the load_model method with that path
model.load_model(model.MODEL_PICKLE_PATH)
```

# Predict

```python
# Getting probabilities for each object
probabilities = model.predict_proba(features)
probabilities
```

```python
# Getting classes
classes = model.predict(features)
classes
```

```python
prediction = model.predict_in_pipeline(features.iloc[[0]])


print('Top classifier')
print(prediction['hierarchical']['top'])

print('\n'*3 + 'Stochastic classifier')
print(prediction['hierarchical']['children']['Stochastic'])

print('\n'*3 + 'Periodic classifier')
print(prediction['hierarchical']['children']['Periodic'])

print('\n'*3 + 'Transient classifier')
print(prediction['hierarchical']['children']['Transient'])

print('\n'*3 + 'Final prediction')
print(prediction['probabilities'])
```
